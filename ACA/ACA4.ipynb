## Visualizing NBA All Star Instagram Relationships with Python

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import SNA data
SNAdata = pd.read_csv("NBA.csv", index_col = 0)
SNAdata
```

The adjacency matrix stands for the Instagram following relationships.
The Rows are the players name, the columns are the tracking result of the following status.
0: Not following
1: Is following

```python
#Visualize the basic directed relationships
G = nx.DiGraph(SNAdata)
plt.figure(figsize=(15,20))
nx.draw_networkx(G,with_labels = True, width = 0.1)
plt.show()
```

From the directed graph, we could see that there are some one-way relationships. For instance, Steph Curry followed Giannis, but Giannis did not follow back. It is possible that Giannis does not use instagram often. He ignored some of the features it has, or he is just not close to any of the players. Domantas has followed many players, however, some of them did not follow back. 

```python
#Visualize the basic relationships
G = nx.Graph(SNAdata)
nx.draw(G, with_labels = True,node_color = 'g')
```

Most of the nodes are well-connected and centered. 
We could tell there are close relationships between the NBA All Star players. 

## Network Statistics 

**Density**
Density is the measurement of how complete the graph is. This is a measure of how complete the graph is (how many edges are present in the network as compared to the total possible number of edges).

```python
#Calculate the density of the edge
density = nx.density(G)

print('The edge density is: %.3f' % (density))
```

This value of 0.536 indivates that this network is well-conneted because we can see more than 40% links among all possible links. 

**Degree Centrality**
This is a common centrality measure, which gives an idea of how 'imporant' each node is in the network. The assumption is that nodes with the most edges are the most important/central as they are directly connected to lots of other nodes.

```python
degree = nx.degree(G)
df = pd.DataFrame(degree)
df
```

From the centrality, we could see that Lebron James has the highest centrality, which means he is the most centralized player throughout the All Star NBA player's Instagram network. Paul Geroge and Damian Lillard are the second.

**Closeness centrality**
Closeness centrality measures "how quickly" a node can travel to the rest of the graph.

```python
close = nx.closeness_centrality(G)
close = pd.DataFrame.from_dict(close, orient = 'index')
close
```

Lebron James could travel fastest to the rest of the graph, following Paul Geroge and Damian Lillard. Julius Randle is the slowest one. 

**Betweenness centrality**

Betweenness centrality is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. The algorithm calculates unweighted shortest paths between all pairs of nodes in a graph. We are using betweenness_centrality() to calculate this metric.

```python
between = nx.betweenness_centrality(G)
between = pd.DataFrame.from_dict(between, orient = 'index')
between
```

Steph Curry has the greatest betweenness centrality overall witha value of 0.05648, followed by Lebron James and Bradley. It mean that Stephen Curry has the greatest amount of influence over the flow of information in the graph. 

**Community detection**

Then let's take a look at the network structure through community detection. This method is very similar to clustering algorithm, which we have seen a few weeks ago. In a social network this is the idea that a large network can be broken down into smaller communinities/cliques. For example, if the network represents the social relationships of all the students at a school, a community/clique would be a friendship group.

The code below calls greedy_modularity_communities() to run the Clauset-Newman-Moore greedy modularity maximization. Greedy modularity maximization begins with each node in its own community and joins the pair of communities that most increases modularity until no such pair exists. We can put the result into a list and see how many communties we have detected.

```python
from networkx.algorithms.community.modularity_max import greedy_modularity_communities

c = list(greedy_modularity_communities(G))

print(len(c))
```

There are 4 commuties we detected. 

```python
#Lets see these 4 clusters
community_0 = sorted(c[0])
community_1 = sorted(c[1])
community_2 = sorted(c[2])
community_3 = sorted(c[3])

print(community_0)
print(community_1)
print(community_2)
print(community_3)
```

Let's get take a look a how to print out the results briefly.

We were using nx.draw() to print out a quick and dirty network, but if you want to draw prettier graphs, you will need to use other functions to plot the network layer by layer.

So first thing is to obtain the positions of each node through spring_layout(). This is just one way of layout.

Then, we will plot each community seperately for different colors.

Also don't forget to plot the edges as well as the labels of the graph.

Finally, use plt.show() to supress the networkx outputs.

```python
# Set seed (optional)
np.random.seed(123)

# Obtain the positions of the nodes
pos = nx.spring_layout(G)

# Print out all the nodes with color
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_0, node_color='green', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_1, node_color='red', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_2, node_color='blue', label = True)
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_3, node_color='yellow', label = True)

# Print out the edges
nx.draw_networkx_edges(G, pos = pos, width = 0.2)

# Make sure the label is printed
nx.draw_networkx_labels(G, pos=pos)

# Show the plot so it supress the other networkx outputs
plt.show()
```
